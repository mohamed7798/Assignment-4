# Task 1: Rectangle class:

```python
class Rectangel:
    def __init__(self,length,width):
        self.length = length
        self.width = width
    def perimeter(self):
        return (2*self.width) + (2*self.length)
    def area(self):
        return (self.width * self.length)
    def display(self):
        print('Length is: ',self.length)
        print('width is: ',self.width)
        print('Perimeter is: ',self.perimeter())
        print('Area is: ',self.area())
class Parallelepipede(Rectangel):
    def __init__(self,length,width,height):
        Rectangel.__init__(self, length, width)
        self.height = height
        def volume(self):
            return (self.length*self.width*self.height)
        def display(self):
            print('Volume is : ',self.volume())
        
```

```python
while True:
    try:
        length = float(input('enter the length of rectangel: '))
        width = float(input('enter the width of rectangel: '))
        r1 = Rectangel(length,width)
        r1.display()
        height = float(input('enter the height of Parallelepipede: '))
        p1 = Parallelepipede(length,width,height)
        p1.display()
        enter = input('if you want to exit type done if you to try again type again: ')
        if enter == 'done':
            break
        else:
            continue
    except Exception as e:
        print(e)
```

# Task 2: Person class and child Student class

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def display(self):
        print("Person name : ", self.name)
        print("Person age = ", self.age)
        print('-'*30)
class Student(Person):
    def __init__(self, name , age , section):
        Person.__init__(self,name, age)
        self.section = section
    #override method
    def display(self):
        print("Student name : ", self.name)
        print("Student age = ", self.age)
        print("Student section = ", self.section)
```

```python
while True:
    try:
        name = input('enter your name : ')
        age = int(input('enter your age : '))
        p1 = Person(name,age)
        p1.display()
        section = input('enter your section : ')
        s1 = Student(name,age,section)
        s1.display()
        enter = input('if you want to exit type done if you to try again type again: ')
        if enter == 'done':
            break
        else:
            continue
    except Exception as e:
        print(e)
```

# Task 3. Computation class 

```python
class Computation:
    def __init__ (self):
        pass
    
    def factorial(self, num):
        temp = 1
        for i in range (1, num + 1):
            temp *=  i
        return print('fact is',temp)
    
    def sum(self, num):
        temp = 1
        for i in range (1, num + 1):
            temp +=  i
        return print('sum is ',temp)
    
    def test_Prim(self, num):
        temp = 0
        for i in range (1, num + 1):
            if (num % i == 0):
                temp +=  1
        if (temp == 2):
            return print('number is prime',True)
        else:
            return print('number is not prime',False)

    def test_prims (self, num1, num2):
        temp = 0
        for i in range (1, num1 + 1):
            if (num1 % i == 0 and num2 % i == 0):
                temp += 1
        if temp == 1:
            print ("The numbers", num1, "and", num2, "are co-primes")
        else:
            print ("The numbers", num1, "and", num2, "are not co-primes")

    def table_Multiplication (self, num):
        for i in range (1,10):
            print (i, "*", num, "=", i * num)
        return print(num)

    def all_Tables(self):
        for k in range (1,10):
            print ("\nthe multiplication table of:", k, "is:")
            for i in range (1,10):
                print (i, "*",  k, "=", i *  k)

    def list_divisors (self, num):
        list_divisors = []
        for i in range (1, num + 1):
            if (num % i == 0):
                list_divisors.append (i)
        return print(list_divisors)

    def list_divisors_Prim (self, num):
        list_divisors = []
        for i in range (1, num + 1):
            if (num % i == 0 and self.test_Prim (i)):
                list_divisors.append (i)
        return print(list_divisors)
```

```python
while True:
    try:
        comput= Computation()
        print('if you want factorial write 1: ')
        print('if you want sum write 2: ')
        print('if you want test Prim write 3: ')
        print('if you want test prims for two numbers write 4: ')
        print('if you want table Multiplication write 5: ')
        print('if you want all Tables write 6: ')
        print('if you want list divisors write 7: ')
        print('if you want list divisors Prim write 8: ')
        ask = input('enter the function you want : ')
        print('-'*30)
        if ask == '1':
            num = int(input('enter the number you want: '))
            print(comput.factorial(num))
            print('-'*30)
        elif ask == '2':
            num = int(input('enter the number you want: '))
            comput.sum(num)
            print('-'*30)
        elif ask == '3':
            num = int(input('enter the number you want: '))
            comput.test_Prim(num)
            print('-'*30)
        elif ask == '4':
            num1 = int(input('enter the first number: '))
            num2 = int(input('enter the seconed number: '))
            comput.test_prims(num1,num2)
            print('-'*30)
        elif ask == '5':
            num = int(input('enter the number you want: '))
            comput.table_Multiplication(num)
            print('-'*30)
        elif ask == '6':
            comput.all_Tables()
            print('-'*30)
        elif ask == '7':
            num = int(input('enter the number you want: '))
            comput.list_divisors(num)
            print('-'*30)
        elif ask == '8':
            num = int(input('enter the number you want: '))
            comput.list_divisors_Prim(num)
            print('-'*30)
        enter = input('if you want to exit type done if you to try again type again: ')
        if enter == 'done':
            break
        else:
            continue
    except Exception as e:
        print(e)
```
